```python
#前期准备
import pandas as pd
data = pd.read_csv('NY property data.csv')
import numpy as np
data['FULLVAL'].replace(0, np.nan, inplace= True)
```

```python
# 辅助set，提取B+B+ZIP
ziptemp=set()
for i in data.index:
    if str(data.at[i,'ZIP']) != 'nan':
        BB = str(data.at[i,'B'])+' '+str(data.at[i,'BLOCK'])+' '+str(int(data.at[i,'ZIP']))
        ziptemp.add(BB)
```

```python
#辅助dict，计数B+B+ZIP：count
ziptemp2={}
for i in ziptemp:
    ziptemp2[i] = 0
for i in data.index:
    if str(data.at[i,'ZIP']) != 'nan':
        BB = str(data.at[i,'B'])+' '+str(data.at[i,'BLOCK'])+' '+str(int(data.at[i,'ZIP']))
        ziptemp2[BB] +=1
```

```python
# 辅助set，ziptemp2中的B+B
ziptemp3 = set()
for i in ziptemp2:
    ziptemp3.add(' '.join(i.split()[0:2]))
```

```python
#ziptemp4 辅助dict   B+B：最大的count
ziptemp4 = {}
ziptemp5 = {}
for i in ziptemp2:
    temp = ' '.join(i.split()[0:2])
    temp2 = i.split()[2]
    if temp not in ziptemp4:
        ziptemp4[temp] = ziptemp2[i]
        ziptemp5[temp] = temp2
    else:
        if ziptemp4[temp] < ziptemp2[i]:
            ziptemp4[temp] = ziptemp2[i]
            ziptemp5[temp] = temp2
```

```python
#ziptemp5即为B+B：ZIP 的字典
ziptemp5
```

```python
for i in data.index:
    if str(data.at[i,'ZIP']) == 'nan':
        BBtemp = str(data.at[i,'B'])+' '+str(int(data.at[i,'BLOCK']))
        limit = 0
        while (BBtemp not in ziptemp5)&(limit<100) :
            BBtemp = str(data.at[i,'B'])+' '+str(int(data.at[i,'BLOCK']+1))#如果ZIP为空的BB在字典中找不到，则参照离该BLOCK最近的BLOCK
            limit+=1
        if limit <100:
            data.at[i,'ZIP'] = int(ziptemp5[BBtemp])
```
