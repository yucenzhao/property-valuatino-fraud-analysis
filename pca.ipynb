```python
import pandas as pd
import numpy as np
data = pd.read_csv('WINNER.csv')
```

```python
data
```

```python
data = data.drop('Unnamed: 0', 1)
data = data.drop('Unnamed: 0.1', 1)
```

```python
data.shape
```

```python
df=pd.DataFrame(data)
```

```python
df=df.ix[:, 46:91]
```

```python
cols = list(df.columns)
df[cols]
```

```python
for col in cols:
    col_zscore = col + '_zscore'
    df[col_zscore] = (df[col] - df[col].mean())/df[col].std(ddof=1)
```

```python
df
```

```python
dfd=df.ix[:,45:90]
dfd
```

```python
dfd.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=45)
pca.fit(dfd)
print (pca.explained_variance_ratio_)
print (pca.explained_variance_)
```

```python
import matplotlib.pyplot as plt
pca = PCA().fit(dfd)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
pca=PCA(n_components=8)
pca.fit(dfd.iloc[:,:45])
```

```python
a = pca.transform(dfd.iloc[:,:45])
```

```python
a
```

```python
a.shape
```

```python
data_pca = pd.DataFrame(a,columns=['pca1', 'pca2', 'pca3', 'pca4', 'pca5', 'pca6', 'pca7', 'pca8'])
```

```python
data_pca
```

```python
colns = list(data_pca)
for coln in colns:
    coln_zscore = coln + '_zscore'
    data_pca[coln_zscore] = (data_pca[coln] - data_pca[coln].mean())/data_pca[coln].std(ddof=1)
```
