```python
import pandas as pd
import numpy as np
data = pd.read_csv('cleaned data.csv')
```

```python
data['lotarea'] = data['LTFRONT'] * data['LTDEPTH']
data['bldarea'] = data['BLDFRONT'] * data['BLDDEPTH']
data['bldvol'] = data['bldarea'] * data['STORIES']
```

```python
for i in data.index:
    data.at[i,'ZIP3'] = int(data.at[i,'ZIP']/100)
```

```python
data['FULL/LOT']=data['FULLVAL']/data['lotarea']
data['FULL/BLD']=data['FULLVAL']/data['bldarea']
data['FULL/STO']=data['FULLVAL']/data['bldvol']
data['AVLAND/LOT']=data['AVLAND']/data['lotarea']
data['AVLAND/BLD']=data['AVLAND']/data['bldarea']
data['AVLAND/STO']=data['AVLAND']/data['bldvol']
data['AVTOT/LOT']=data['AVTOT']/data['lotarea']
data['AVTOT/BLD']=data['AVTOT']/data['bldarea']
data['AVTOT/STO']=data['AVTOT']/data['bldvol']
```

```python
data['FL/LT/Z5'] = data.groupby(['ZIP'])['FULL/LOT'].apply(lambda x: x/x.mean())
data['FL/BL/Z5'] = data.groupby(['ZIP'])['FULL/BLD'].apply(lambda x: x/x.mean())
data['FL/ST/Z5'] = data.groupby(['ZIP'])['FULL/STO'].apply(lambda x: x/x.mean())
data['AL/LT/Z5'] = data.groupby(['ZIP'])['AVLAND/LOT'].apply(lambda x: x/x.mean())
data['AL/BL/Z5'] = data.groupby(['ZIP'])['AVLAND/BLD'].apply(lambda x: x/x.mean())
data['AL/ST/Z5'] = data.groupby(['ZIP'])['AVLAND/STO'].apply(lambda x: x/x.mean())
data['AT/LT/Z5'] = data.groupby(['ZIP'])['AVTOT/LOT'].apply(lambda x: x/x.mean())
data['AT/BL/Z5'] = data.groupby(['ZIP'])['AVTOT/BLD'].apply(lambda x: x/x.mean())
data['AT/ST/Z5'] = data.groupby(['ZIP'])['AVTOT/STO'].apply(lambda x: x/x.mean())
```

```python
data['FL/LT/Z3'] = data.groupby(['ZIP3'])['FULL/LOT'].apply(lambda x: x/x.mean())
data['FL/BLD/Z3'] = data.groupby(['ZIP3'])['FULL/BLD'].apply(lambda x: x/x.mean())
data['FL/STO/Z3'] = data.groupby(['ZIP3'])['FULL/STO'].apply(lambda x: x/x.mean())
data['AL/LT/Z3'] = data.groupby(['ZIP3'])['AVLAND/LOT'].apply(lambda x: x/x.mean())
data['AL/BLD/Z3'] = data.groupby(['ZIP3'])['AVLAND/BLD'].apply(lambda x: x/x.mean())
data['AL/STO/Z3'] = data.groupby(['ZIP3'])['AVLAND/STO'].apply(lambda x: x/x.mean())
data['AT/LT/Z3'] = data.groupby(['ZIP3'])['AVTOT/LOT'].apply(lambda x: x/x.mean())
data['AT/BLD/Z3'] = data.groupby(['ZIP3'])['AVTOT/BLD'].apply(lambda x: x/x.mean())
data['AT/STO/Z3'] = data.groupby(['ZIP3'])['AVTOT/LOT'].apply(lambda x: x/x.mean())
```

```python
data['FL/LT/TX'] = data.groupby(['TAXCLASS'])['FULL/LOT'].apply(lambda x: x/x.mean())
data['FL/BL/TX'] = data.groupby(['TAXCLASS'])['FULL/BLD'].apply(lambda x: x/x.mean())
data['FL/ST/TX'] = data.groupby(['TAXCLASS'])['FULL/STO'].apply(lambda x: x/x.mean())
data['AL/LT/TX'] = data.groupby(['TAXCLASS'])['AVLAND/LOT'].apply(lambda x: x/x.mean())
data['AL/BL/TX'] = data.groupby(['TAXCLASS'])['AVLAND/BLD'].apply(lambda x: x/x.mean())
data['AL/ST/TX'] = data.groupby(['TAXCLASS'])['AVLAND/STO'].apply(lambda x: x/x.mean())
data['AT/LT/TX'] = data.groupby(['TAXCLASS'])['AVTOT/LOT'].apply(lambda x: x/x.mean())
data['AT/BL/TX'] = data.groupby(['TAXCLASS'])['AVTOT/BLD'].apply(lambda x: x/x.mean())
data['AT/ST/TX'] = data.groupby(['TAXCLASS'])['AVTOT/STO'].apply(lambda x: x/x.mean())
```

```python
data['FL/LT/B'] = data.groupby(['B'])['FULL/LOT'].apply(lambda x: x/x.mean())
data['FL/BL/B'] = data.groupby(['B'])['FULL/BLD'].apply(lambda x: x/x.mean())
data['FL/ST/B'] = data.groupby(['B'])['FULL/STO'].apply(lambda x: x/x.mean())
data['AL/LT/B'] = data.groupby(['B'])['AVLAND/LOT'].apply(lambda x: x/x.mean())
data['AL/BL/B'] = data.groupby(['B'])['AVLAND/BLD'].apply(lambda x: x/x.mean())
data['AL/ST/B'] = data.groupby(['B'])['AVLAND/STO'].apply(lambda x: x/x.mean())
data['AT/LT/B'] = data.groupby(['B'])['AVTOT/LOT'].apply(lambda x: x/x.mean())
data['AT/BL/B'] = data.groupby(['B'])['AVTOT/BLD'].apply(lambda x: x/x.mean())
data['AT/ST/B'] = data.groupby(['B'])['AVTOT/STO'].apply(lambda x: x/x.mean())
```

```python
data['FL/LT/all'] = data.groupby(['ZIP','ZIP3','TAXCLASS','B'])['FULL/LOT'].apply(lambda x: x/x.mean())
data['FL/BLD/all'] = data.groupby(['ZIP','ZIP3','TAXCLASS','B'])['FULL/BLD'].apply(lambda x: x/x.mean())
data['FL/STO/all'] = data.groupby(['ZIP','ZIP3','TAXCLASS','B'])['FULL/STO'].apply(lambda x: x/x.mean())
data['AL/LT/all'] = data.groupby(['ZIP','ZIP3','TAXCLASS','B'])['AVLAND/LOT'].apply(lambda x: x/x.mean())
data['AL/BLD/all'] = data.groupby(['ZIP','ZIP3','TAXCLASS','B'])['AVLAND/BLD'].apply(lambda x: x/x.mean())
data['AL/STO/all'] = data.groupby(['ZIP','ZIP3','TAXCLASS','B'])['AVLAND/STO'].apply(lambda x: x/x.mean())
data['AT/LT/all'] = data.groupby(['ZIP','ZIP3','TAXCLASS','B'])['AVTOT/LOT'].apply(lambda x: x/x.mean())
data['AT/BLD/all'] = data.groupby(['ZIP','ZIP3','TAXCLASS','B'])['AVTOT/BLD'].apply(lambda x: x/x.mean())
data['AT/STO/all'] = data.groupby(['ZIP','ZIP3','TAXCLASS','B'])['AVTOT/LOT'].apply(lambda x: x/x.mean())
```
